```python
people = pd.read_csv('/Users/andrejalekseevic/study/pytthon/5/titanic_with_labels.csv', index_col=0, sep=' ')
people
```

```python
people = people.dropna(subset=['sex'])
people = people[~people['sex'].isin(['-', 'Не указан', 'не указан'])]
people['sex'] = people['sex'].replace({'м': 1, 'ж': 0, 'М': 1, 'Ж': 0, 'M': 1, 'Мужчина': 1})

people['row_number'].fillna(people['row_number'].max(), inplace=True)

people['liters_drunk'] = people['liters_drunk'].mask(people['liters_drunk'] < 0, people['liters_drunk'].mean())

q_low = people['liters_drunk'].quantile(0.25)
q_hi = people['liters_drunk'].quantile(0.75)
q_range = q_hi - q_low

people['liters_drunk'] = people['liters_drunk'].mask(
    (people['liters_drunk'] < q_hi + 1.5 * q_range) & (people['liters_drunk'] > q_low - 1.5 * q_range),
    people['liters_drunk'].mean()
)

people
```

```python
kid = [
    (people['age'] < 18),
    (people['age'] >= 18)
]
choices = [1, 0]

people['age_kid'] = np.select(kid, choices)

adult = [
    (people['age'] > 18 ) & (people['age'] < 50),
    (people['age'] <= 18)
]
choices = [1, 0]

people['age_adult'] = np.select(adult, choices)

elderly = [
    (people['age'] > 50),
    (people['age'] <= 50)
]
choices = [1, 0]

people['age_elderly'] = np.select(elderly, choices)

del people['age']

people
```

```python
people['drink'] = people['drink'].replace({'Cola': 0, 'Fanta': 0, 'Beerbeer': 1, 'Bugbeer': 1, 'Water': 0, 'Strong beer': 1, 'Наше пиво': 1})

people
```

```python
sessions = pd.read_csv('/Users/andrejalekseevic/study/pytthon/5/cinema_sessions.csv', index_col=0, sep=' ')

sessions
```

```python
people = pd.merge(people, sessions, left_on="check_number", right_on="check_number")

people['session_start'] = pd.to_timedelta(people['session_start'])

mt = pd.to_timedelta('12:00:00.000')
dt = pd.to_timedelta('20:00:00.000')
ev = pd.to_timedelta('23:59:59.999')
```

```python
morning = [
    (people['session_start'] < mt ) & (people['session_start'] > ev),
    (people['session_start'] >= mt ) & (people['session_start'] <= ev),
]
choices = [1, 0]

people['morning'] = np.select(morning, choices)

day = [
    (people['session_start'] < dt ) & (people['session_start'] > mt),
    (people['session_start'] >= dt ) & (people['session_start'] <= mt),
]
choices = [1, 0]

people['day'] = np.select(day, choices)

evening = [
    (people['session_start'] < ev) & (people['session_start'] > dt),
    (people['session_start'] >= ev) & (people['session_start'] <= dt),
]
choices = [1, 0]

people['evening'] = np.select(evening, choices)

del people['session_start']

people
```
